```python
#memphis versus nashville
#double bar chart nashville & memphis
#percent of people with dogs, percent with cats, percent with kids
```

```python
bar_categories = ["% with dogs", "% with cats", "% with kids"]
memphis_values = []
nashville_values = []
```

```python
condition_memphis = df["city"] == "Memphis"
condition_nashville = df["city"] == "Nashville"
```

```python
kid_owners_crit = df['Presence_of_Children_v2'] == "Y"
```

```python
Nmemphis = len(df[condition_memphis])
Nnashville = len(df[condition_nashville])
```

```python
Ndog_owners_memphis = len(df[dog_owners_crit & condition_memphis])
Ndog_owners_nashville = len(df[dog_owners_crit & condition_nashville])
```

```python
percent_dog_owners_memphis = Ndog_owners_memphis/Nmemphis
percent_dog_owners_nashville = Ndog_owners_nashville/Nnashville
```

```python
#repeat for cats
Ncat_owners_memphis = len(df[cat_owners_crit & condition_memphis])
Ncat_owners_nashville = len(df[cat_owners_crit & condition_nashville])
percent_cat_owners_memphis = Ncat_owners_memphis/Nmemphis
percent_cat_owners_nashville = Ncat_owners_nashville/Nnashville
```

```python
#repeat for kids
Nkid_owners_memphis = len(df[kid_owners_crit & condition_memphis])
Nkid_owners_nashville = len(df[kid_owners_crit & condition_nashville])
percent_kid_owners_memphis = Nkid_owners_memphis/Nmemphis
percent_kid_owners_nashville = Nkid_owners_nashville/Nnashville
```

```python
#organize data into arrays
memphis_values = [percent_dog_owners_memphis, percent_cat_owners_memphis, percent_kid_owners_memphis]
nashville_values = [percent_dog_owners_nashville, percent_cat_owners_nashville, percent_kid_owners_nashville]
```

```python
#double bar chart
X_axis = np.arange(len(bar_categories))

plt.bar(X_axis - 0.2, memphis_values, 0.4, label = 'Memphis', color = '#a2cccc')
plt.bar(X_axis + 0.2, nashville_values, 0.4, label = 'Nashville', color = '#b21818')

plt.xticks(X_axis, bar_categories)
#plt.xlabel("Groups")
plt.ylabel("Percent of Population")
plt.title("Dog, Cat, Child Ownership in Nashville & Memphis")
plt.legend()
```
