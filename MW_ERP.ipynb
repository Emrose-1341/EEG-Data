```python
# Step 1: Create the DataFrame
folder_path = '/content/drive/MyDrive/WM Text File/WM Text Files (1)/WM Text Files'
WM_df = []

for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
      try:
          with open(file_path, 'r', encoding = 'latin-1') as file:
            content = file.read()
      except:
          with open(file_path, 'r', encoding = 'utf-8') as file:
            content = file.read()

      #Lexical_df.append({f"Content of {filename}:\n{content}\n"})
      WM_df.append({'filename': filename, 'content': content})

ERP_data = pd.DataFrame(WM_df)

#WM_df.append({'filename': filename, 'content': content})
#ERP_data = pd.DataFrame(WM_df)
#print("Column names:", ERP_data.apply.columns.tolist())
```

```python
keyword = 'Average:'
averages = []

for index, row in ERP_data.iterrows():
    lines = row['content'].split('\n')
    average_value = None
    for line in lines:
        if keyword in line:
            average_value = line.split(keyword)[-1].strip()  # Extract part after 'Average:' and remove leading/trailing spaces
            break  # Stop searching for 'Average:' once found

    averages.append(average_value)

ERP_data['Average'] = averages

#averages = pd.DataFrame(ERP_data['Average'])   # Create a new column 'Average' in the DataFrame
```
